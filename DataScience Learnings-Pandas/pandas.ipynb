```python
###MAKING CHANGES TO IT
#df.head(5)
# 1) to get total col: df['Total']= df['HP']+df['Attack']+df['Defense']+df['Sp. Atk']+df['Sp. Def']+df['Speed']
# drops total column: df=df.drop(columns=['Total'])
#to add a col: df['Total']=df.iloc[:,4:10].sum(axis=1) and it adds 4th to 9th col axis=0 adds vertically
df['Total']=df.iloc[:,4:10].sum(axis=1)
cols=list(df.columns.values)
df=df[cols[0:4]+[cols[-1]]+cols[4:12]] #this is to shuffle the column the way user wants..column total has been shifted to 4
df.head(5)
```

```python
##Filtering data
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison')] #AND condition
df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')] #OR condition
new_df=df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
#new_df.to_csv('filtered.csv')
new_df=new_df.reset_index(drop=True)
new_df
new_df.to_csv('filtered.csv')
df.loc[~df['Name'].str.contains('Mega')]#filters out and removes name called mega
```

```python
##Complicated filtering
import re
df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]
#to ignore the uppercase lowercase mistake
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```
