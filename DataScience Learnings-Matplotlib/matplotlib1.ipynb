```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

####Basic Graph



```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
#Resize your graph(dpi specifies pixels per inch.When saving probably should use 300 if possible)
plt.figure(figsize=(2,2),dpi=100)
#plt.plot(x,y,label='2x',color='red',linewidth=4,marker='*',linestyle="--",markersize=7,markeredgecolor='blue')
#Line 1
##shorthand notation
#fmt = '[color][marker][line]'
plt.plot(x,y,'r.--',label='2x')




#Line 2
#select interval we want to plot points at
x2=np.arange(0,4.5,0.5)

#plot part of the graph as line
plt.plot(x2[:6],x2[:6]**2,label='x^2')

#plot remainder of graph
plt.plot(x2[5:],x2[5:]**2,'b--')

#Add a title (specify font parameters with fontdict)
plt.title('Our first graph',fontdict={'fontname': 'Comic Sans MS','fontsize':20})

#x and y labels
plt.xlabel('X axis',fontdict={'fontname': 'Arial','fontsize':17})
plt.ylabel('Y axis',fontdict={'fontname': 'Arial','fontsize':17})

#scale of your graph X,Y tickmarks
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#Add a legend
plt.legend()

#save figure(dpi 300 is good when saving so graph has a high resolution)
plt.savefig('mygraph.png',dpi=300)

#show plot
plt.show()
```
